# 5. Data Normalization

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df['DEP_DELAY'] = scaler.fit_transform(df[['DEP_DELAY']])
df['ARR_DELAY'] = scaler.fit_transform(df[['ARR_DELAY']])
```

```python
df.head()
```

```python
# Central Tendency

mean = df['DEP_DELAY'].mean()
median = df['DEP_DELAY'].median()
mode = df['DEP_DELAY'].mode()[0]

print("For Deperature Delay\n")
print(f"Mean: {mean}, Median: {median}, Mode: {mode}\n\n")

mean = df['ARR_DELAY'].mean()
median = df['ARR_DELAY'].median()
mode = df['ARR_DELAY'].mode()[0]

print("For Arrival Delay\n")
print(f"Mean: {mean}, Median: {median}, Mode: {mode}\n")
```

```python
# Dispersion

variance = df['DEP_DELAY'].var()
std_dev = df['DEP_DELAY'].std()
data_range = df['DEP_DELAY'].max() - df['DEP_DELAY'].min()

print("For Deperature Delay\n")
print(f"Variance: {variance}, Standard Deviation: {std_dev}, Range: {data_range}\n\n")

variance = df['ARR_DELAY'].var()
std_dev = df['ARR_DELAY'].std()
data_range = df['ARR_DELAY'].max() - df['ARR_DELAY'].min()

print("For Arrival Delay\n")
print(f"Variance: {variance}, Standard Deviation: {std_dev}, Range: {data_range}")
```

```python
from sklearn.preprocessing import LabelEncoder

categorical_cols = df.select_dtypes(include=['object']).columns

# Create a LabelEncoder instance
le = LabelEncoder()

# Convert each object column to numeric using Label Encoding
for column in categorical_cols:
    df[column] = le.fit_transform(df[column])
```

```python
from sklearn.metrics import roc_curve, auc

models = {
    'Decision Tree': DecisionTreeClassifier(),
    'Logistic Regression': LogisticRegression(),
    'KNN': KNeighborsClassifier(),
    'Random Forest': RandomForestClassifier(),
    'XGBoost': XGBClassifier(),
    'Bagging Classifier': BaggingClassifier(),
    'Gradient Boosting Classifier':GradientBoostingClassifier(),
    'Naive Bayes': GaussianNB(),
    'Perceptron':Perceptron(),
    'Neural Network' : MLPClassifier()
    
}
for model_name, model in models.items():
    model.fit(x_train, y_train)
plt.figure(figsize=(10, 8))

for model_name, model in models.items():
    # Compute predicted probabilities or decision scores
    if hasattr(model, "predict_proba"):  # For models with predict_proba
        y_score = model.predict_proba(x_test)[:, 1]
    else:  # For models with decision_function (e.g., SVM)
        y_score = model.decision_function(x_test)
    
    # Calculate FPR, TPR, and AUC
    fpr, tpr, _ = roc_curve(y_test, y_score)
    roc_auc = auc(fpr, tpr)
    
    # Plot ROC curve for the model
    plt.plot(fpr, tpr, lw=2, label=f'{model_name} (AUC = {roc_auc:.2f})')

# Add diagonal line for random guessing
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--', label='Random Guessing')

# Customize plot aesthetics
plt.xlabel('False Positive Rate', fontsize=12)
plt.ylabel('True Positive Rate', fontsize=12)
plt.title('ROC Curve Comparison', fontsize=14)
plt.legend(loc="lower right", fontsize=10)
plt.grid(alpha=0.3)
plt.show()
```
