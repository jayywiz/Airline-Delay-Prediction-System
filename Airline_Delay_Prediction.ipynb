# 5. Data Normalization

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df['DEP_DELAY'] = scaler.fit_transform(df[['DEP_DELAY']])
df['ARR_DELAY'] = scaler.fit_transform(df[['ARR_DELAY']])
```

```python
df.head()
```

```python
# Central Tendency

mean = df['DEP_DELAY'].mean()
median = df['DEP_DELAY'].median()
mode = df['DEP_DELAY'].mode()[0]

print("For Deperature Delay\n")
print(f"Mean: {mean}, Median: {median}, Mode: {mode}\n\n")

mean = df['ARR_DELAY'].mean()
median = df['ARR_DELAY'].median()
mode = df['ARR_DELAY'].mode()[0]

print("For Arrival Delay\n")
print(f"Mean: {mean}, Median: {median}, Mode: {mode}\n")
```

```python
# Dispersion

variance = df['DEP_DELAY'].var()
std_dev = df['DEP_DELAY'].std()
data_range = df['DEP_DELAY'].max() - df['DEP_DELAY'].min()

print("For Deperature Delay\n")
print(f"Variance: {variance}, Standard Deviation: {std_dev}, Range: {data_range}\n\n")

variance = df['ARR_DELAY'].var()
std_dev = df['ARR_DELAY'].std()
data_range = df['ARR_DELAY'].max() - df['ARR_DELAY'].min()

print("For Arrival Delay\n")
print(f"Variance: {variance}, Standard Deviation: {std_dev}, Range: {data_range}")
```

# 6. Data Visualizations

```python
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
#1. Count Plot

plt.figure(figsize=(10, 6))
sns.countplot(x='CARRIER', data=df)
plt.title('Count Plot: Flight Count')
plt.show()
```

```python
# 2. Bar Plot

df_grouped = df.groupby('YEAR')[['DEP_DELAY', 'ARR_DELAY']].sum().reset_index()

plt.figure(figsize=(12, 6))

# Plot the total departure delays
plt.subplot(1, 2, 1)
sns.barplot(x='YEAR', y='DEP_DELAY', data=df_grouped, color='blue')
plt.title('Total Departure Delays greater than 15 min')
plt.xlabel('Year')
plt.ylabel('Total Departure Delays')

# Plot the total arrival delays
plt.subplot(1, 2, 2)
sns.barplot(x='YEAR', y='ARR_DELAY', data=df_grouped, color='orange')
plt.title('Total Arrival Delays greater than 15 min')
plt.xlabel('Year')
plt.ylabel('Total Arrival Delays')

plt.tight_layout()
plt.show()
```

```python
# 3. Line plot

df['FL_DATE'] = pd.to_datetime(df['FL_DATE'])
df_sorted = df.sort_values('FL_DATE')

df_sorted['Cumulative Departure Delay'] = df_sorted['DEP_DELAY_NEW'].cumsum()
df_sorted['Cumulative Arrival Delay'] = df_sorted['ARR_DELAY_NEW'].cumsum()

plt.figure(figsize=(10, 6))

# Lineplot for cumulative departure delays
sns.lineplot(x='FL_DATE', y='Cumulative Departure Delay', data=df_sorted, label='Departure Delay', color='blue')

# Lineplot for cumulative arrival delays
sns.lineplot(x='FL_DATE', y='Cumulative Arrival Delay', data=df_sorted, label='Arrival Delay', color='orange')

plt.title('Line Plot: Cumulative Flight Delays')
plt.xlabel('Flight Date')
plt.ylabel('Cumulative Delay (Minutes)')

plt.legend()
plt.show()
```

```python
# 4. Boxplot

plt.figure(figsize=(12, 6))
sns.boxplot(x='CARRIER', y='DEP_DELAY_NEW', data=df)
plt.title('Boxplot: Departure Delays')
plt.xticks(rotation=45)
plt.show()
```

```python
# 5. Q-Q plot

import scipy.stats as stats

plt.figure(figsize=(12, 6))

# Subplot for Departure Delay
plt.subplot(1, 2, 1)
stats.probplot(df['DEP_DELAY_NEW'].dropna(), dist="norm", plot=plt)
plt.title('Q-Q Plot for Departure Delays')

# Subplot for Arrival Delay
plt.subplot(1, 2, 2)
stats.probplot(df['ARR_DELAY_NEW'].dropna(), dist="norm", plot=plt)
plt.title('Q-Q Plot for Arrival Delays')

plt.tight_layout()
plt.show()
```

```python
# 6. Pie chart

plt.figure(figsize=(8, 8))
cancellation_status = df['CANCELLED'].value_counts()
plt.pie(cancellation_status, labels=cancellation_status.index, autopct='%1.1f%%', startangle=90)
plt.title('Pie chart: Flight Cancellation Status')
plt.show()
```

```python
# 7. Heatmap of correlation between numerical features

plt.figure(figsize=(12, 8))
corr_matrix = df[numerical_columns].corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Heatmap')
plt.show()
```

```python
# 8. Pairplot

sns.pairplot(df[['DEP_DELAY_NEW', 'ARR_DELAY_NEW', 'CANCELLED', 'DIVERTED']])
plt.show()
```

```python
# 9. Scatterplot

plt.figure(figsize=(10, 6))
sns.scatterplot(x='DEP_DELAY_NEW', y='ARR_DELAY_NEW', data=df)
plt.title('Scatterplot: Departure vs Arrival Delays')
plt.show()
```

```python
# 10. Bar plot of Origin Cities

plt.figure(figsize=(12, 6))
top_origins = df['ORIGIN_CITY'].value_counts().nlargest(10)
sns.barplot(x=top_origins.index, y=top_origins.values)
plt.xticks(rotation=45)
plt.title('Top 10 Origin Cities')
plt.show()
```

```python
# 11. Histogram

plt.figure(figsize=(10, 6))

plt.subplot(1, 2, 1)
sns.histplot(df['DEP_DELAY'], bins=30, kde=True)
plt.title('Distribution of Departure Delays')

plt.subplot(1, 2, 2)
sns.histplot(df['ARR_DELAY'], bins=30, kde=True)
plt.title('Distribution of Arrival Delays')
plt.show()
```

```python
# 12. Violin Plot

plt.figure(figsize=(10, 6))
sns.violinplot(x='CARRIER', y='DIVERTED', data=df)
plt.title('Violin Plot: Flight Diverted across Carriers')
plt.show()
```

```python
# 13. Strip Plot

plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.stripplot(x='YEAR', y='DEP_DELAY_NEW', data=df, jitter=True)
plt.title('Departure Delay per Year')

plt.subplot(1, 2, 2)
sns.stripplot(x='YEAR', y='ARR_DELAY_NEW', data=df, jitter=True)
plt.title('Arrival Delay per Year')
plt.show()
```

# 7. Data Transformation

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler

numerical_cols = df.select_dtypes(include=['float64', 'int64']).columns
categorical_cols = df.select_dtypes(include=['object']).columns

# Standardize numerical features
scaler = StandardScaler()
df_scaled_numerical = pd.DataFrame(scaler.fit_transform(df[numerical_cols]), columns=numerical_cols)

# One-Hot Encode categorical features
encoder = OneHotEncoder(handle_unknown='ignore')
df_encoded_categorical = pd.DataFrame(encoder.fit_transform(df[categorical_cols]))

# Combine numerical and categorical features
df_combined = pd.concat([df_scaled_numerical, df_encoded_categorical], axis=1)

```
